```python
import pandas as pd
from collections import Counter
```

```python
df = pd.read_csv("reviews_cleaned.csv")
```

```python
comments = list(df["comments_cleaned"][0:100])
```

```python
words = []
for line in comments:
    for word in line.split():
       words.append(word)
```

```python
len(words)
```

```python
word_counts = Counter(words)
```

```python
#sorted_vocab = sorted(word_counts, key=word_counts.get, reverse=True)
```

```python
sorted_vocab = sorted(word_counts.items(), key=lambda x:x[1], reverse=True)
```

```python
for e in enumerate(sorted_vocab):
    print(e)
```

```python
{ii: word for ii, word in enumerate(sorted_vocab)}
```

```python
int_to_vocab = {ii: word for ii, word in enumerate(sorted_vocab)}
```

```python
vocab_to_int = {word: ii for ii, word in int_to_vocab.items()}
```
