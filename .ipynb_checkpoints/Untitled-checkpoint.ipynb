```python
import pandas as pd
import spacy
nlp = spacy.load('en')
doc1 = nlp(u'Hello hi there!')
doc2 = nlp(u'Hello hi there!')
doc3 = nlp(u'Hey whatsup?')

print(doc1.similarity(doc2)) # 0.999999954642
print(doc2.similarity(doc3)) # 0.699032527716
print(doc1.similarity(doc3)) # 0.699032527716
```

```python
reviews = pd.read_csv("reviews_cleaned.csv")
```

```python
reviews.head()
```

```python
import pandas as pd
import spacy
nlp = spacy.load('en')
```

```python
with open("comparison.txt", "w") as fp:
    for i in reviews.iloc.index:        
        doc_i = nlp(reviews.iloc[i]["comments_cleaned"])
        for j in reviews.iloc.index:
            doc_j = nlp(reviews.iloc[j]["comments_cleaned"])
            txt = str(i) + "," + str(j) + "," + str(doc_i.similarity(doc_j))               
            fp.write(txt + "\n")
```
